```python
data = pd.read_csv(r"C:\Users\Dell\Downloads\Employee-Attrition - Employee-Attrition (1).csv")
```

```python
log_cols = data.select_dtypes(include = 'number').columns
for col in log_cols:
    data[col] = np.log10(data[col] + 1)
    
```
